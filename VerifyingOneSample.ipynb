```python
# Define the global variables
arrOfStartCells = []
arrOfEndCells = []
arrOfCells = []
arrOfCells1 = []
game_id = 0
num_steps = 0
```

```python
# Define the functions we will use
def DisplayMatrix(arr):
    print("|-----------------------------------------------------------|")
    for row in range(25):
        row_str = "|"
        for col in range(25):
            row_str += " "
            row_str += str(arr[row][col])
            if (col+1)%5 == 0:
                row_str += " |"
        print(row_str)
        if (row+1)%5 == 0:
            print("|-----------------------------------------------------------|")

    

def CopyCells(arr1,arr2):
    for row in range(25):
        for col in range(25):
            arr1[row][col] = arr2[row][col]
            
def CompareCells(arr1,arr2):
    for row in range(25):
        for col in range(25):
            if (arr1[row][col] != arr2[row][col]):
                return False
    return True

def CountNumberOfOnes(arr):
    count = 0
    for row in range(25):
        linecount = 0
        for col in range(25):
            if (arr[row][col] == 1):
                count += 1
                linecount += 1
    return count

def CountNumberOfZeros(arr):
    count = 0
    for row in range(25):
        for col in range(25):
            if (arr[row][col] == 0):
                count += 1
    return count

    
    
def CountNumberOfMismatches(arr1,arr2):
    count = 0
    for row in range(25):
        for col in range(25):
            if (arr1[row][col] != arr2[row][col]):
                count += 1
    return count

def PrintNumbers(pre_str,arr):
    count0 = 0
    count1 = 0
    for row in range(25):
        for col in range(25):
            if (arr[row][col] == 0):
                count0 += 1
            if (arr[row][col] == 1):
                count1 += 1
    print("matrix ",pre_str,"has ",count0, "zeros, and ",count1," ones")

def InitMatrices(arr,arrSt,arrEnd,arrTemp1,arrTemp2):
    global game_id
    global num_steps
    
    game_id = int(arr[0])
    num_steps = int(arr[1])

    cellNumber = 2
    for row in range(25):
        arrSt.append([])
        for col in range(25):
            arrSt[row].append(0)
            arrSt[row][col] = int(arr[cellNumber])
            cellNumber += 1
            
    for row in range(25):
        arrEnd.append([])
        for col in range(25):
            arrEnd[row].append(0)
            arrEnd[row][col] = int(arr[cellNumber])
            cellNumber += 1

    for row in range(25):
        arrTemp1.append([])
        arrTemp2.append([])
        for col in range(25):
            arrTemp1[row].append(0)
            arrTemp2[row].append(0)
            arrTemp1[row][col] = 0
            arrTemp2[row][col] = 0
    
    
    
def RunOneStep(arrStartStep,arrEndStep):
    for nRow in range(25):
        for nCol in range(25):
            # Compute number of neighbors
            # Each cell has exactly 8 neighbors
            nNumNgbs = int(0)

            #Left-upper
            nVirtualRow = nRow-1
            nVirtualCol = nCol-1
            if (nVirtualRow == -1):
                nVirtualRow = 24
            if (nVirtualCol == -1):
                nVirtualCol = 24
                
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]


            #Directly above
            nVirtualRow = nRow-1
            nVirtualCol = nCol
            if (nVirtualRow == -1):
                nVirtualRow = 24
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            #Right-upper
            nVirtualRow = nRow-1
            nVirtualCol = nCol+1
            if (nVirtualRow == -1):
                nVirtualRow = 24
            if (nVirtualCol == 25):
                nVirtualCol = 0
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            #Directly right
            nVirtualRow = nRow
            nVirtualCol = nCol+1
            if (nVirtualCol == 25):
                nVirtualCol = 0
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            #Right-lower
            nVirtualRow = nRow+1
            nVirtualCol = nCol+1
            if (nVirtualRow == 25):
                nVirtualRow = 0
            if (nVirtualCol == 25):
                nVirtualCol = 0
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]


            #Directly below
            nVirtualRow = nRow+1
            nVirtualCol = nCol
            if (nVirtualRow == 25):
                nVirtualRow = 0
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            #Left-below
            nVirtualRow = nRow+1
            nVirtualCol = nCol-1
            if (nVirtualRow == 25):
                nVirtualRow = 0
            if (nVirtualCol == -1):
                nVirtualCol = 24
		
            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            #Directly left
            nVirtualRow = nRow
            nVirtualCol = nCol-1
            if (nVirtualCol == -1):
                nVirtualCol = 24

            nNumNgbs += arrStartStep[nVirtualRow][nVirtualCol]

            # Implement the rules for the Game-of-Life
            if (arrStartStep[nRow][nCol] == 1):
                # This is the case we have a square with a living cell.
                # It dies if it have 4 or more neighbors OR
                # if it has 0 or 1 neighbors
                if (nNumNgbs > 3 or nNumNgbs < 2):
                    arrEndStep[nRow][nCol] = 0
                else:
                    arrEndStep[nRow][nCol] = 1
            else:
                # this is the case the square has NO living cell.
                # It "reproduces" if it has exactly 3 neighbors
                if (nNumNgbs == 3):
                    arrEndStep[nRow][nCol] = 1
                else:
                    arrEndStep[nRow][nCol] = 0
			
#end def RunOneStep(arrStartStep,arrEndStep):
```

```python
# Get the filename containing one sample
strfilename = input("Input the filename that contains the sample: ")
print("The filename = ",strfilename)

with open(strfilename) as file:
    data = file.readlines()

data = data[0]  # Get the 0th line in the file

arr = []
arr = data.split(sep=',')

# First entry is the Game-ID
game_id = int(arr[0])
print("Game ID = ",game_id)

# Second entry is the number of steps it takes to go from Start-Square to End-Square
num_steps = int(arr[1])
print("Number of Steps = ",num_steps)
```

```python
# Initialize the Start-Square and the End-Square from the file entered above
InitMatrices(arr,arrOfStartCells,arrOfEndCells,arrOfCells,arrOfCells1)
print("\n\n Start Matrix")
DisplayMatrix(arrOfStartCells)
print("\n\n End Matrix")
DisplayMatrix(arrOfEndCells)
```

```python
# Run the number of steps required to go from "Start-Square" to "End-Square"

CopyCells(arrOfCells,arrOfStartCells)

for i in range(num_steps):
    RunOneStep(arrOfCells,arrOfCells1)
    CopyCells(arrOfCells,arrOfCells1)

    print("\nHere's the matrix on completing ",i+1," step")
    DisplayMatrix(arrOfCells)
```

```python
# Compare the square we obtained after running the required number of steps, 
# compare that to the "End-Square"
theCompare = CompareCells(arrOfCells,arrOfEndCells)
print("After {}-steps, the match on End-Game is ".format(num_steps),theCompare)
if theCompare == True:
    print("\n* * * * * S U C C E S S * * * * *")
else:
    print("\n* * * * * F A I L U R E * * * * *")
    
```
